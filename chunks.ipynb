Helper function for finding inactive users

```python
def helper(matrix):
    toss = []
    for index, row in matrix.iterrows():
        res = np.sum(row[1:])
        if res<=15:
            # toss.append(int(row[0]))
            toss.append(index)
    return toss
            
```

```python
userID = userTime.loc[:,"Unnamed: 0"]
pre2002 = userTime.iloc[:, 1:782 :1]
pre2002.index = np.array(userID)
display(pre2002)
startingUsers = []
for index, row in pre2002.iterrows():
    res = np.sum(row[1:])
    if res>0:
        # toss.append(int(row[0]))
        startingUsers.append(index)
print(len(startingUsers))
```

```python
users = open("users.txt","r")
toss = set()
for line in users:
    user = int(line.replace(",\n",""))
    if user in startingUsers:
        pass
    else:
        toss.add(user)
```
