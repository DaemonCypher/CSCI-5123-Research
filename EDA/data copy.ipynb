```python
import datetime
from dateutil import parser
import pandas as pd
import numpy as np
from datetime import date, timedelta
```

```python
# MERGED data
# From https://www.tutorialspoint.com/How-to-merge-multiple-files-into-a-new-file-using-Python
file_names = ["combined_data_1.txt", "combined_data_2.txt", "combined_data_3.txt", "combined_data_4.txt"]

with open("data.txt", "a") as outfile:
    for fname in file_names:
        with open(fname, "r") as infile:
            for line in infile:
                outfile.write(line)
```

```python
# finding the start and end of temporal data
small = datetime.datetime(2002, 1, 1)
big =  datetime.datetime(2002, 1, 1)


with open("data.txt") as infile:
    for line in infile:
        data = line.strip().split(',')
        if len(data) > 2:
            # From https://www.geeksforgeeks.org/nlp-using-dateutil-to-parse-dates/
            DT = parser.parse(data[2])
            if DT < small:
                small =DT
            if DT> big:
                big = DT
print(big)
print(small)

# 2005-12-31 00:00:00
# 1999-11-11 00:00:00
```

```python
# get number of times user has rated an item
user={}
with open("data.txt") as infile:
    for line in infile:
        data = line.strip().split(',')
        if len(data) > 2:
            if data[0] not in user:
                user[data[0]]=1
            else:
                user[data[0]]+=1
            
```

```python
# filter for any user with more than 100 ratings
# From https://stackoverflow.com/questions/8425046/the-best-way-to-filter-a-dictionary-in-python
filter = {k: v for k, v in user.items() if v >= 100}
# number of users who have more than 100 ratings

print(len(filter))
```

```python
# creating a user x time matrix
user_time={}
for users in filter:
    user_time[users]=[]

with open("data.txt") as infile:
    for line in infile:
        data = line.strip().split(',')
        if len(data) > 2:
            user =data[0]
            if user in filter:
                user_time[user].append(data[2])
            
```

```python
# create columns for user-time matrix
# From https://stackoverflow.com/questions/7274267/print-all-day-dates-between-two-dates
start_date = date(1999, 11, 11) 
end_date = date(2005, 12, 31)    

delta = end_date - start_date   

days =[]
for i in range(delta.days + 1):
    day = start_date + timedelta(days=i)
    days.append(str(day))
```

```python
# load users, dont need this line if you have the users already defined elsewhere
netflixUserFile = open("users.txt","r")
netflixUser =[]
for line in netflixUserFile:
    res =line.replace(",\n","")
    netflixUser.append(res)
# From https://www.geeksforgeeks.org/different-ways-to-create-pandas-dataframe/
userTime = pd.DataFrame(0, index=netflixUser, columns=days)
```
