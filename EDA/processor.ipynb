```python
import pandas as pd
```

```python
# files = ['combined_data_1.csv', 'combined_data_2.csv', 'combined_data_3.csv', 'combined_data_4.csv']
# df =pd.concat([pd.read_csv(file) for file in files], ignore_index=True)
df = pd.read_csv('Netflix_User_Ratings.csv')
df
```

```python
movie_import_before_2002 = df.groupby('movie_id').agg({'timestamp': 'min'}).reset_index()
movie_import_before_2002['timestamp'] = pd.to_datetime(movie_import_before_2002['timestamp'])
movie_import_before_2002 = movie_import_before_2002[movie_import_before_2002['timestamp'] < '2002-01-01']
movie_import_before_2002
```

```python
#only keep movies that were imported before 2002
df = df[df['movie_id'].isin(movie_import_before_2002['movie_id'])]
df
```

```python
user_stats = df.groupby('user_id')['timestamp'].agg(['min', 'max']).reset_index()
user_stats['min'] = pd.to_datetime(user_stats['min'])
user_stats['max'] = pd.to_datetime(user_stats['max'])
user_stats
```

```python
valid_users = user_stats[(user_stats['min'] < '2002-01-01') & (user_stats['max'] >= '2005-01-01')]
valid_users
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df_filter = df[df['timestamp']>= '2002-01-01']
df_filter
```

```python
df_filter['slice'] = ((df['timestamp'].dt.year - 2002) * 12 + (df['timestamp'].dt.month - 1)) // 3 + 1
df_filter
```

```python
df_filter = df_filter[df_filter['user_id'].isin(valid_users['user_id'])]
df_filter
```

```python
valid_users['count'] = df_filter.groupby('user_id')['movie_id'].size().values
valid_users
```

```python
user_slice_count = df_filter.groupby(['user_id', 'slice']).size().unstack(fill_value=0)
user_slice_count
```

```python
user_slice_count['valid_slice'] = user_slice_count[user_slice_count >= 15].count(axis=1)
user_slice_count
```

```python
active_users = valid_users[valid_users['count'] >= 100]
active_users = active_users.merge(user_slice_count['valid_slice'], on='user_id', how='left')
active_users = active_users[active_users['user_id'].isin(user_slice_count[user_slice_count['valid_slice'] >= 4].index)]
active_users
```

```python
#convert to csv
active_users.to_csv('active_users.csv', index=False)
```

```python
user_slice_count['count'] = user_slice_count.sum(axis=1) - user_slice_count['valid_slice']
user_slice_count
```
