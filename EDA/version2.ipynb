```python
def helper(matrix):
    toss = []
    for index, row in matrix.iterrows():
        res = np.sum(row[1:])
        if res<15:
            toss.append(index)
    return toss
```

```python
gran = 91
toss = set()
for i in range(0,13):
    start = chunks[i][0]
    end = chunks[i][1]
    splice = startingSplice.iloc[:, start:end :1]
    res = helper(splice)
    for i in res:
        toss.add(i)
    print(start,end,len(startingSplice)-len(toss))
```
